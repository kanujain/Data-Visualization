```python
import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('E:\\2020\\Quarantine\\Python\\tips.csv')
data.head()
```

```python
data.groupby('time').mean()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data['time'], data['total_bill'])  
plt.title('Time verses Total prise paid')
plt.show()
```

```python
avg_tip = data['tip'].mean()

data['above_avg'] = (data['tip'] - avg_tip) > 0
data['above_avg']
```

```python
sns.countplot('time', data=data, 
              hue='above_avg',
              order = data['time'].value_counts().index)  
                                                        # Bill in each time gave above avg tip / not

plt.title('Bills and Tips')
plt.show()
```

```python
sns.countplot('time', data=data, 
              hue='sex',order = data['time'].value_counts().index)  
                                                        # Bill in each time gave above avg tip / not

plt.title('Gender wrt Tips')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot('sex', 'total_bill', data=data, palette='nipy_spectral')

plt.title('Gender wrt Tips')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot('size', 'total_bill', data=data, palette='nipy_spectral')

plt.title('Size of partyb V/S Total Bill')
plt.show()
```

```python
#Catagorical with plot kind

plt.figure(figsize=(10,10))
sns.catplot('size', 'total_bill', data=data,
            hue= 'sex',kind='bar',palette='nipy_spectral')

plt.title('Size of party V/S Total Bill WRT SEX')
plt.show()

plt.figure(figsize=(10,10))
sns.catplot('size', 'total_bill', data=data,
            hue= 'time',kind='bar',palette='nipy_spectral')

plt.title('Size of party V/S Total Bill WRT TIME')
plt.show()
```

```python
#Good visualisation of Categorical value in scatter plot form

plt.figure(figsize=(10,10))
sns.stripplot('size', 'total_bill', data=data)

plt.title('Size of party V/S Total Bill')
plt.show()
```

```python
#To visualize each and every data point we use swarm Plot

plt.figure(figsize=(10,10))
sns.swarmplot('size', 'total_bill', data=data)

plt.title('Size of party V/S Total Bill')
plt.show()
```

```python
#Swarm plot using hue

plt.figure(figsize=(10,10))
sns.swarmplot('size', 'total_bill', data=data, hue='above_avg', dodge=True)  
                                    #dodge is used to seperate the color representation

plt.title('Size of party V/S Total Bill WRT Avg Tip')
plt.show()
```
