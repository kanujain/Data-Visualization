```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('E:\\2020\\Quarantine\\Python\\Automobile_data.csv',na_values='?') 
                                        #na_values places NaN to places that are not available
data.head()
```

```python
data.shape
#Check how many null values exist in column
data.isnull().sum()
#dropping NaN values
data.dropna(inplace=True)
data.shape
```

```python
data.to_csv('E:\\2020\\Quarantine\\Python\\Automobile_PROCESSED_data.csv', index=False)
```

```python
data.columns
```

```python
plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.distplot(data['price'], color='b', bins=50)  #Bins is the number of bars in the graph
plt.title('Automobile Price distribution')
plt.show()
```

```python
#If we donot want to see the Hostogram only the KDE curve 

plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.distplot(data['price'], color='b', hist=False)  #Bins is the number of bars in the graph
plt.title('Automobile Price distribution')
plt.show()

#RUG Plot: It represents individual data points with small lines 

plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.distplot(data['price'], color='b', hist=False, rug=True)  #Bins is the number of bars in the graph
plt.title('Automobile Price distribution')
plt.show()
```

```python
# If we only want rug plot

plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.rugplot(data['price'], color='b', height=0.3)  #Bins is the number of bars in the graph
plt.title('Automobile Price distribution')
plt.show()
```

```python
#Just the kde plot 

plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.kdeplot(data['price'], color='b', shade=True)  #Bins is the number of bars in the graph
plt.title('Automobile Price distribution')
plt.show()
```

```python
#bIVARIATE 
    #Scatter plot
    
plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.scatterplot(x=data['horsepower'],y=data['price'], color='b', s=100)  #Bins is the number of bars in the graph
                                                                        #s is for the size of dots
plt.title('HorsePower of Automobile verses Price of Automobile')
plt.show()
```

```python
plt.figure(figsize=(12,8))
            #Distribution Plot
            #For UNIVARIATE Data
sns.scatterplot(x=data['horsepower'],y=data['price'], color='b',hue=data['num-of-cylinders'], s=100)  #Bins is the number of bars in the graph
                                                                        #s is for the size of dots
                                                                        #
plt.title('HorsePower of Automobile verses Price of Automobile')
plt.show()
```

```python
#Regression plot
```

```python
#This tells whether linear relationship exist b/w variables or not
plt.figure(figsize=(12,8))
            #Regression Plot
            #For UNIVARIATE Data
sns.regplot(x=data['horsepower'],y=data['price'])  
                            #Provides regression line with confidence interval
plt.title('HorsePower of Automobile verses Price of Automobile')
plt.show()
```

```python
plt.figure(figsize=(12,8))
            #Regression Plot
            #For UNIVARIATE Data
sns.regplot(x=data['highway-mpg'],y=data['price'])  
                            # Here we see negar=tive relationship
                            #Bz high horse power cars have low milage and are very expensive
plt.title('highway-mpg of Automobile verses Price of Automobile')
plt.show()
```

```python
sns.jointplot(x=data['highway-mpg'],y=data['price']) #We get the frequency distribution of each variable
plt.show()

sns.jointplot(x=data['highway-mpg'],y=data['price'], kind = 'reg')
plt.show()
```

```python
sns.jointplot(x=data['highway-mpg'],y=data['price'], kind = 'kde')  #KDE joint plot
```

```python
sns.jointplot(x=data['highway-mpg'],y=data['price'], kind = 'hex')  #Hex joint bin plot
```

```python
fig, ax = plt.subplots(figsize=(6,6))

sns.kdeplot(data['highway-mpg'],data['price'], ax=ax)
sns.rugplot(data['highway-mpg'], color = 'limegreen', ax=ax)
sns.rugplot(data['price'],vertical=True,color='red',ax=ax)

plt.show()
```
